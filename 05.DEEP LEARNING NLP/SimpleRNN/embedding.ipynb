```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
#sentences 
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good'
]
```

```python
sent
```

```python
## define the vocabulary size
voc_size = 10000
```

```python
#3One hot representation

one_hot_repre = [one_hot(words,voc_size) for words in sent]
one_hot_repre
```

```python
#word embedding representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.preprocessing.sequence import pad_sequences #sent length 
from tensorflow.keras.models import Sequential
```

```python
import numpy as np
```

```python
sent_length = 8
embedded_doc = pad_sequences(one_hot_repre,padding='pre',maxlen=sent_length)
embedded_doc
```

```python
## feature representation
dim = 10
```

```python
model = Sequential()
model.add(Embedding(voc_size,dim,input_length =sent_length))
model.compile('adam','mse')
```

```python
model.summary()
```

```python
model.predict(embedded_doc)
```
